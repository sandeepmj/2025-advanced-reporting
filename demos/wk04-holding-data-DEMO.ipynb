## ```Dictionaries```

A Python ```Dictionary```, or ```dict``` is another common and powerful datatype.

Think of it as a collection of data points about a single element.

For example, a car model:

- brand: Audi
- body: SUV
- model: Q4
- trim: Leather
- cost: 60_000
- on_sale: False
- vin: 2973
- type: EV

A dictionary would hold this data as **key: value** pairs

A dictionary is wrapped in ```{}```.

Each key: value pair is separated by a comma.

The format is ```{"key1_string": "string_value", "key2_number": 12, "key3_bolean": True}```

````python
## build AND CALL the car_1 dictionary
car_1 = {"brand": "Audi",
         "body": "SUV",
        "model": "Q4",
        "trim": "leather",
        "cost": 60_000,
        "on_sale": False,
        "VIN": "2990",
        "type": "EV"}

car_1

````

Build me a dictionary called car_2 with the following data points.

Note: In this case the vin is first. The order in a dictionary doesn't matter.

- vin: 8303
- brand: Honda
- body: Sedan
- model: Accord
- type: gas
- trim: Cloth
- cost: 42,000
- on sale: yes

````python
## build and call car_2
car_2 = {"VIN": "8303",
        "brand": "Honda",
        "body": "Sedan",
        "model": "Accord",
        "type": "gas",
        "trim": "cloth",
        "cost": 42_000,
        "on_sale": True}
car_2
````

### A ```list``` can holds ```dicts```

````python
## create a list called cars_list here
cars_list = [car_1, car_2]

cars_list
````

## Why is a ```list``` that holds ```dicts``` even remotely of relevance to us?

````python
## JUST RUN TO DEMO
## NO NEED FOR STUDENTS TO TYPE
df = pd.DataFrame(cars_list)
df.to_csv("cars.csv", encoding = "UTF-8", index = False)
````

## But what if we want a single item, or a range of items from this list?

#### To retrieve a data point we can call it's index position

```list_name[index_position]```

````python
## what position is PVC?
toxic_chemicals.index("PVC")
````

````python
toxic_chemicals
````

### ZERO always comes first!
The first value in every programming language is ```zero-indexed```, or the count always starts at ```zero```.



````python
## what position is 'Polyvinyl Chloride'
toxic_chemicals.index('PFOA')
````

#### But usually we want to retrieve data, not the index position

You retrieve data by calling the list, plus an index position:

```list_name[index_position]```

````python
## get the first item
toxic_chemicals[0]
````

````python
## get the third item and store in a variable:
tox2 = toxic_chemicals[2]
tox2
````

````python
## what type of data is this single item sliced off a list?
type(tox2)
````

````python
## call the last item in the list
toxic_chemicals[-1]
````

````python
## call the penultimate item in the list
toxic_chemicals[-2]
````
