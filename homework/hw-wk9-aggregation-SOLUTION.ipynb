### A few problems:

1. Too many decimal places and scientific notation.
2. Way too many columns to be meanginful or easy to read.


Deal with the first problem here:

```python
## Run the code that resets the default display here to round the numbers, remove scientific notation, etc.
pd.options.display.float_format = '{:,.0f}'.format
```

```python
## Run the code here to return all summary statistics in one go
df.describe()
```

### Deal with the second problem of having too much summary data:

```python
## What is the sum of all the forgiven loans
df["ForgivenessAmount"].sum()
```

```python
## What is the sum of all the forgiven loans
## format with dollar sign and thousands separators
## round to zero decimals
f'${df["ForgivenessAmount"].sum():_.0F}'
```

```python
## store in variable called total_loan_forgiven
total_loan_forgiven = df["ForgivenessAmount"].sum()
total_loan_forgiven
```

```python
## f string format that variable to 2 decimal places
f"${total_loan_forgiven:_.2F}"
```

```python
## return the median for CurrentApprovalAmount loans, InitialApprovalAmount loans and ForgivenessAmount loan amounts
df[["InitialApprovalAmount", "CurrentApprovalAmount", "ForgivenessAmount"]].median()
```

```python
##  Return the above as a frame
(df[["InitialApprovalAmount", "CurrentApprovalAmount", "ForgivenessAmount"]].median()).to_frame("Median")
```

```python
## get the count, mean and median for CurrentApprovalAmount
df["CurrentApprovalAmount"].agg(["count", "mean", "median"])
```

```python
## automatically return a frame for the count, mean and median for InitialApprovalAmount, CurrentApprovalAmount
## note the double brackets because we are passing a list of columns - Automatic to frame
df[["InitialApprovalAmount", "CurrentApprovalAmount"]].agg(["count", "mean", "median"])
```
